### Features and Feature Types

```python
#checking data format and null count
data.info()
```

```python
#Visualising the absolute and relative amount of missing values in each feature
missing_counts = data.isna().sum()
missing_percent = (missing_counts / len(data)) * 100

missing_df = pd.DataFrame({
    'Missing Values': missing_counts.values,
    'Percent Missing': missing_percent.values
}, index=missing_counts.index)

missing_df = missing_df.sort_values(by='Percent Missing', ascending=False)

missing_df
```

Preleminary conclusions:
- Seems like the only 2 columns that **do not** contain missing values are those relating to the customer ID and to the target variable (which explicits if the given person will donate)
- Some columns, like *WEALTH_RATING*, *DONOR_AGE*, *INCOME_GROUP* present a rate of missing values close or even higher than 25%, maybe requiring imputation or removal
- There are several groups of columns presenting the exact same number of missing values, suggesting that these features might be related and missing not at random.
